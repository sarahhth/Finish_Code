```python
import pandas as pd
import plotly.express as px  # Impor Plotly Express

# Membaca dataset
file_path = r"C:\MATA KULIAH\AKADEMIK\SEMESTER 6\Data Science\week 6\IrisDataset.csv"
datasetIris = pd.read_csv(file_path)



# Konversi species ke kategori
datasetIris['species'] = datasetIris['species'].astype('category')

# Mengubah kategori species menjadi kode numerik
datasetIris['species_code'] = datasetIris['species'].cat.codes

#  Gunakan nama kolom yang benar dari dataset
fig = px.parallel_coordinates(datasetIris,
    dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
    color='species_code',  # Gunakan kode numerik untuk warna
    color_continuous_scale=px.colors.diverging.Tealrose
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px
from sklearn.preprocessing import MinMaxScaler

# Path dataset
file_path = r"C:\Users\user\Data Science\Week 6\preprocessing1.csv"

df = pd.read_csv(file_path)

# Cek beberapa baris pertama
print(df.head())

# Cek nama kolom
print(df.columns)
```

```python
selected_columns = ['Durasi Laptop', 'Edukasi Youtube di laptop', 'Penggunaan Medsos di Laptop',
                    'Durasi Ponsel', 'Nilai US Matematika']
```

```python
df_selected = df[selected_columns].copy()
```

```python
df_selected = df_selected.fillna(df_selected.mean())
```

```python
scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df_selected), columns=df_selected.columns)
```

```python
#Bagaimana Durasi Laptop, Edukasi YouTube, Penggunaan Medsos, Durasi Ponsel memengaruhi Nilai US Matematika.
fig = px.parallel_coordinates(df_normalized, 
                              dimensions=df_normalized.columns,
                              title="Parallel Coordinates: Durasi Belajar vs Prestasi Akademik",
                              color=df_normalized["Nilai US Matematika"],  # Pewarnaan berdasarkan nilai matematika
                              color_continuous_scale=px.colors.sequential.Viridis)

# 6. Tampilkan plot
fig.show()
```
